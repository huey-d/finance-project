```python
import pandas as pd
import alpha_vantage.fundamentaldata
import psycopg2

import keys
from keys import api_key

pd.set_option('display.max_columns', None)  # set option to display all columns
```

```python
def get_fundamental_data(ticker: str = 'MSFT'):
    fd = alpha_vantage.fundamentaldata.FundamentalData(key=api_key)

    q_is = fd.get_income_statement_quarterly(symbol=ticker)
    is_df = pd.DataFrame.from_records(q_is[0])

    q_cf = fd.get_cash_flow_quarterly(symbol=ticker)
    cf_df = pd.DataFrame.from_records(q_cf[0])
    
    return is_df, cf_df
```

```python
is_df, cf_df = get_fundamental_data()
```
