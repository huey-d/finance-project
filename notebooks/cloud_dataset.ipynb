```python
import pandas as pd
import numpy as np
import sqlalchemy

import pandas as pd
import keys

import os

pd.set_option('display.max_columns', None)  # set option to display all columns
pd.options.display.max_colwidth = 1000
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
engine = sqlalchemy.create_engine(keys.cloud_connection_string)
```

```python
historical_data_query = """
    SELECT * FROM historical_data ORDER BY date ASC
"""

microsoft_query = """
    SELECT * FROM microsoft
"""

balance_sheet_data_query = """
SELECT * FROM "balance_sheet_data" ORDER BY "fiscalDateEnding" ASC
"""

income_statement_data_query = """
    SELECT * FROM "income_statement_data" ORDER BY "fiscalDateEnding" ASC
"""

cash_flow_data_query = """
    SELECT * FROM cash_flow_data ORDER BY "fiscalDateEnding" ASC
"""
```

```python
stock_df =  pd.read_sql(historical_data_query, engine)
stock_df
```

```python
def save_dataframe_to_csv(dataframe, file_name: str):
    """
    Saves a pandas dataframe to a CSV file.

    Args:
        dataframe (pandas.DataFrame): The dataframe to save.
        file_path (str): The path to the file to save (including the filename and extension).
    """
    dataframe.to_csv(file_name, index=False)
```

```python
save_dataframe_to_csv(stock_df, 'stock_data.csv')
save_dataframe_to_csv(overview_df, 'microsoft.csv')
save_dataframe_to_csv(balance_sheet_df, 'balance_sheet.csv')
save_dataframe_to_csv(income_statement_df, 'income_statement.csv')
save_dataframe_to_csv(cash_flow_df, 'cash_flow.csv')
```
